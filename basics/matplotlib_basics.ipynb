# Matplotlib Basics 

## 1. Introduction

In this notebook, we will cover the basics of Matplotlib, a powerful library for creating static, animated, and interactive visualizations in Python. We'll focus on using matplotlib.pyplot to create various types of plots.

## 2. Importing Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 3. Basic Line Plot

```python
# Generating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a basic line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

## 4. Scatter Plot

```python
# Generating sample data
x = np.random.rand(100)
y = np.random.rand(100)

# Creating a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

## 5. Bar Plot

```python
# Generating sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

# Creating a bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='blue')
plt.title("Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## 6. Histogram

```python
# Generating sample data
data = np.random.randn(1000)

# Creating a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## 7. Pie Chart

```python
# Generating sample data
sizes = [20, 30, 25, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# Creating a pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart")
plt.axis('equal')
plt.show()
```

## 8. Subplots

```python
# Generating sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
plt.figure(figsize=(12, 6))

# First subplot
plt.subplot(1, 2, 1)
plt.plot(x, y1, label="Sine", color='blue')
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(x, y2, label="Cosine", color='red')
plt.title("Cosine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.tight_layout()
plt.show()
```

## 9. Saving a Plot

```python
# Generating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating and saving a plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Sine Wave", color='purple')
plt.title("Saving a Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)

# Save the figure as a PNG file
plt.savefig('data/sine_wave_plot.png')

plt.show()
```

## 10. Conclusion

In this notebook, we explored the basics of Matplotlib's `pyplot` module. We created various types of plots, including line plots, scatter plots, bar plots, histograms, pie charts, and subplots. We also learned how to customize these plots and save them to files. Matplotlib is a versatile tool for creating a wide range of visualizations in Python.
