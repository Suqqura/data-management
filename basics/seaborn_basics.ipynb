# Seaborn Basics

## 1. Introduction
In this notebook, we'll cover the basics of the Seaborn library for data visualization. Seaborn is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics.

## 2. Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Set a default aesthetic for Seaborn
sns.set(style="whitegrid")
```

## 3. Loading a Sample Dataset

```python
# Load the tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

## 4. Basic Plots

```python
# Scatter plot showing the relationship between total bill and tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Histogram of total bill amounts
sns.histplot(tips['total_bill'], kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
# Box plot showing total bill amounts by day of the week
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

## 5. Advanced Plots

```python
# Pair plot to see relationships between all numeric variables
sns.pairplot(tips)
plt.suptitle("Pair Plot of Tips Dataset", y=1.02)
plt.show()
```

```python
# Correlation heatmap
# Select only numeric columns
numeric_columns = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr = numeric_columns.corr()

# Create the heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Facet grid of scatter plots of total bill vs tip, separated by smoker status
g = sns.FacetGrid(tips, col="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()

# Adjust the layout manually
plt.suptitle("Facet Grid of Total Bill vs Tip by Smoker Status")
plt.show()
```

## 6. Customizing Plots

```python
# Customizing plot aesthetics
sns.set(style="darkgrid")

# Custom color palette
sns.set_palette("pastel")

# Example plot with customizations
sns.scatterplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
plt.title("Customized Scatter Plot with Seaborn")
plt.show()
```

## 7. Saving Plots

```python
# Save a plot to a file
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.savefig('data/box_plot_total_bill_by_day.png')
plt.show()
```

## 8. Conclusion
In this notebook, we explored the basics of Seaborn for data visualization. We created various types of plots including scatter plots, histograms, box plots, and heatmaps. We also learned how to customize Seaborn plots and save them to files. Seaborn is a powerful tool for visualizing data and exploring relationships between variables.
